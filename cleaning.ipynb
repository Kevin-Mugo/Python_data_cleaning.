Arrange the column "Contract" into the same format.

```python
def get_contract(contract):
    if contract == 'Free' or 'On Loan' in contract:
        start_date = np.nan
        end_date = np.nan
        contract_length = 0
    else:
        start_date , end_date = contract.split(' ~ ')
        start_year = int(start_date[:4])
        end_year = int(end_date[:4])
        contract_length = (end_year-start_year)
    return start_date , end_date , contract_length

#Calling the function above to the "Contract column" and creating new columns.

new_columns = ['Contract Start' , 'Contract End' ,'Contract Length (Years)']
new_data = dfcopy['Contract'].apply(lambda x : pd.Series(get_contract(x)))


for i in range(len(new_columns)):
    dfcopy.insert(loc=dfcopy.columns.get_loc('Contract')+1+i , column= new_columns[i] , value = new_data[i])
```

```python
pd.set_option('display.max_columns' , None)
dfcopy.head()
```

```python
dfcopy[['Contract','Contract Start' , 'Contract End' ,'Contract Length (Years)']].sample(50)
```

```python
#Contract categories.

def contract_status(contract):
    if contract == 'Free':
        return "Free"
    elif 'On Loan' in contract:
        return 'On Loan'
    else:
        return "On Contract"

dfcopy.insert(dfcopy.columns.get_loc('Contract Length (Years)')+1 , 'Contract status' , dfcopy['Contract'].apply(contract_status))

dfcopy.sample(5)
```

WEIGHT : Player's weights in kilograms(Kgs).

```python
dfcopy['Weight'].dtypes
```

```python
#As seen above , the weight column is an object rather than intergers , lets correct that.
#Check for unique values.
dfcopy['Weight'].unique()
```

```python
#Convert lbs to kgs and remove "kg" from expressions to make them intergers.
def weight_conversion(weight):
    if "kg" in weight:
        return int(weight.strip("kg"))
    else:
        int_lbs = int(weight.strip('lbs'))

        #converting lbs  into kgs.
        return round(int_lbs/2.205)

#apply the function to the height column.
dfcopy['Weight'] = dfcopy['Weight'].apply(weight_conversion)
dfcopy['Weight'].unique()
```

```python
#Remane the column.
dfcopy = dfcopy.rename(columns = {'Weight': "Weight(Kgs)"})
dfcopy.sample(1)
```

```python
def hitchange(h):
    if isinstance(h, str) and "K" in h:
        h = h.replace('K', "")
        h = int(float(h) * 1000)
        return round(h)
    else: 
        return h

#apply the function
dfcopy['Hits'] = dfcopy['Hits'].apply(hitchange)


dfcopy['Hits'].unique()
```
